```python
import numpy as np
import matplotlib.pyplot as plt

# 样本数据：面积（平方英尺）和房价（美元）
X = np.array([1000, 1500, 2000, 2500, 3000])
y = np.array([189785, 254562, 335200, 389610, 400000])

# 特征缩放
X_mean = np.mean(X)
X_std = np.std(X)
X_normalized = (X - X_mean) / X_std

# 增加x0=1（截距）
X_b = np.c_[np.ones(X_normalized.shape[0]), X_normalized]

# 参数初始化
theta = np.zeros(X_b.shape[1])

# 超参数
learning_rate = 0.01
iterations = 1000
theta
```

```python

# 损失函数（均方误差）
def compute_cost(X, y, theta):
    m = len(y)
    return (1 / (2 * m)) * np.sum((X.dot(theta) - y) ** 2)

# 梯度下降算法
def gradient_descent(X, y, theta, learning_rate, iterations):
    m = len(y)
    cost_history = np.zeros(iterations)
    
    for i in range(iterations):
        gradient = (1 / m) * X.T.dot(X.dot(theta) - y)
        print(gradient,theta)
        theta -= learning_rate * gradient
        cost_history[i] = compute_cost(X, y, theta)
    
    return theta, cost_history

# 训练模型
theta, cost_history = gradient_descent(X_b, y, theta, learning_rate, iterations)

# 预测
def predict(X, theta):
    return X.dot(theta)

# 可视化损失函数的变化
plt.figure(figsize=(12, 6))
plt.plot(range(iterations), cost_history, 'b')
plt.xlabel('iteration')
plt.ylabel('loss')
plt.title('损失函数值随迭代次数的变化')
plt.show()
```

```python

# 可视化数据点和回归直线
plt.figure(figsize=(12, 6))
plt.scatter(X, y, color='red', label='训练数据')
x_values = np.linspace(min(X), max(X), 100)
x_values_normalized = (x_values - X_mean) / X_std
x_values_b = np.c_[np.ones(x_values_normalized.shape[0]), x_values_normalized]
y_values = predict(x_values_b, theta)
plt.plot(x_values, y_values, color='blue', label='回归直线')
plt.xlabel('面积（平方英尺）')
plt.ylabel('房价（美元）')
plt.title('房价与房屋面积的回归分析')
plt.legend()
plt.show()

# 示例预测：预测面积为1200平方英尺的房价
area = 1200
area_normalized = (area - X_mean) / X_std
area_with_intercept = np.array([1, area_normalized])
predicted_price = predict(area_with_intercept, theta)

print(f"预测面积为 {area} 平方英尺的房价: ${predicted_price:.2f}")
```
